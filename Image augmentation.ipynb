```python
from tensorflow.keras.preprocessing import image
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img=image.load_img('cat.78.jpg',target_size=(200,200))
```

```python
img
```

```python
type(img)
```

```python
datagen=ImageDataGenerator(
          rotation_range=30,
          shear_range=0.2,
          zoom_range=0.2,
          horizontal_flip=True,
          width_shift_range=0.2,
         height_shift_range=0.2
                 )
```

```python
img=image.img_to_array(img)
```

```python
img.shape
```

```python
type(img)
```

```python
type(datagen)
```

```python
input_batch=img.reshape(1,200,200,3)
```

```python
#flow only for 1 image in a batch,for all images flow_from_directory
i=0
for op in datagen.flow(input_batch,batch_size=1,save_to_dir='aug'):
    i=i+1
    if i==10:
        break
```
